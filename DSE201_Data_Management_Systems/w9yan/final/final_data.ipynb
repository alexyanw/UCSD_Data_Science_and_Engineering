```python
#Create entries for customer table customer name = State_code+number(0:10000)
team_list = []
team_table = []
TEAM_SIZE = 500

for i in range (TEAM_SIZE):
        team_list.append('team'+str(i))
        team_info = {
            "name" : 'team'+str(i),
            "coach": 'coach'+str(i)
        }
        team_table.append(team_info)

team_table_t = tuple(team_table)
```

```python
match_table = []
match_list = []
team_undefeated = 18

for h in set(numpy.random.randint(TEAM_SIZE,size=TEAM_SIZE)):
    for v in (set(range(0,TEAM_SIZE))-set([h])):
        match_info = {
            "hteam"  : 'team'+str(h),
            "vteam"  : 'team'+str(v),
            "hscore" : randint(0,99),
            "vscore" : randint(0,99)
        }
        if h == team_undefeated:
            match_info['vscore'] = min(match_info['hscore'] - 10, 0)
        if v == team_undefeated:
            match_info['hscore'] = min(match_info['vscore'] - 10, 0)
        match_table.append(match_info)

match_table_t = tuple(match_table)
```
